el total de combinaciones teniendo en cuenta las restricciones son:
$$V^{5}_{9} · P_{4} = \frac{9!}{(9-5)!} · 4! = 9! = 362.880 $$

```
numeros = ["1","2","3","4","5","6","7","8","9"]
operadores = ["+","-","*","/"]

def todos_operadores_usados(solucion):
   return "+" in solucion and "-" in solucion and "*" in solucion and "/" in solucion
  

def es_solucion(solucion, N):
  return todos_operadores_usados(solucion) and len("".join(solucion)) == 9 and eval("".join(solucion)) == N  
    
def fuerza_bruta(N):

  for n1 in numeros:
    for o1 in operadores:
      for n2 in numeros:
        for o2 in operadores:
          for n3 in numeros:
            for o3 in operadores:
              for n4 in numeros:
                for o4 in operadores:
                  for n5 in numeros:
                    sol = [str(n1), str(o2), str(n2), str(o2), str(n3), str(o3), str(n4), str(o4), str(n5)]
                    if es_solucion(sol, N):
                      return sol
fuerza_bruta(4)
```

```
numeros = ["1","2","3","4","5","6","7","8","9"]
operadores = ["+","-","*","/"]

# Funcion que devuelve True si la solucione s prometedora o no (es decir, puede desembocar en una solucion al problema)
def es_prometedora(solucion,etapa):
  sol = "".join(solucion)
  return len(set(sol)) == len(sol)
    
# Esta funcion decide si se han utilizado todos los operadores en la solucion suministrada (es una restriccion del problema)
def todos_operadores_usados(solucion):
   return "+" in solucion and "-" in solucion and "*" in solucion and "/" in solucion
  
# Esta funcion decide si la solucion suministrada es una solucion al problema o no
def es_solucion(solucion, N):
  return todos_operadores_usados(solucion) and len("".join(solucion)) == 9 and eval("".join(solucion)) == N  
      
# Funcion principal, los parametros son: N (numero que queremos obtener), solucion (variable donde se guardan las soluciones intermedias y la final) 
# y etapa (variable que establece en que etapa nos encontramos del algoritmo)
def operaciones(N,solucion=[],etapa=0): 
  #print(f"etapa : {etapa}, solucion : {solucion}")
  
  if len(solucion) == 0:
    solucion = ["","" ,"" ,"" ,"" ,"" ,"" ,"",""]
    
  if not etapa %2:
    for numero in numeros :
      solucion[etapa] = numero
      if es_prometedora(solucion,etapa):
        if es_solucion(solucion, N):
          print(f"\nLa solucion es {solucion}")

        elif etapa < 8:
          operaciones(N,solucion,etapa+1)
          solucion[etapa+1] =""
            
  else:
    for operador in operadores:
      solucion[etapa] = operador
      if es_prometedora(solucion,etapa):
        if es_solucion(solucion, N):
          print(f"\nLa solucion es {solucion}")

        elif etapa < 8:
          operaciones(N,solucion,etapa+1)
          solucion[etapa+1] =""

          
```
